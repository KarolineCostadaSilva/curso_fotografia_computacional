#Referência: https://opencv24-python-tutorials.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_geometric_transformations/py_geometric_transformations.html

```python
import cv2
from google.colab.patches import cv2_imshow
# import numpy as np


im_path = "./imgcourse/Morangos.bmp"
image = cv2.imread(im_path)

#Redução em 25% para cada dimensão
Fx = 0.25; Fy = 0.25
image_resize = cv2.resize(image,  None, fx = Fx, fy = Fy, interpolation = cv2.INTER_CUBIC)


cv2_imshow(image)
print("\n\n")
cv2_imshow(image_resize)

#Ampliação em 50% para cada dimensão
Fx = 1.5; Fy = 1.5
image_resize = cv2.resize(image,  None, fx = Fx, fy = Fy, interpolation = cv2.INTER_CUBIC)
print("\n\n")
cv2_imshow(image_resize)
```
