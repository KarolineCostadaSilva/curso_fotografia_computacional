```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

im_path = "./imgcourse/Cin.jpg"
image = cv2.imread(im_path)

#As dimensões da imagem são extraídas
height, width = image.shape[:2]

#Os deslocamentos são estabelecidos em 1/4 para cada dimensão
quarter_height, quarter_width = height/4, width/4

#A nova imagem transladada é calculada
T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])
img_translation = cv2.warpAffine(image, T, (width, height))

cv2_imshow(image)
print("\n\n")
cv2_imshow(img_translation)
Flag = cv2.imwrite('Translation1.jpg', img_translation)

#Novos deslocamentos são calculados em direções opostas
quarter_height, quarter_width = -height/4, -width/4

#Uma nova imagem transladada é calculada
T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])
img_translation = cv2.warpAffine(image, T, (width, height))

print("\n\n")
cv2_imshow(img_translation)
Flag = cv2.imwrite('Translation2.jpg', img_translation)




```
