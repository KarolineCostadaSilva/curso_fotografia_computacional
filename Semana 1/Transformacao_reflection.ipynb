# Código em Python usando dados do repositório

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

im_path = "./imgcourse/Cin.jpg"
img = cv2.imread(im_path)


# Mirror in x direction (flip horizontally)
imgX = np.flip(img, axis=1)
# imgX = imgX = img[:, ::-1, :]

# Mirror in y direction (flip vertically)
imgY = np.flip(img, axis=0)
# imgY = img[::-1, :, :]

# Mirror in both directions (flip horizontally and vertically)
imgXY = np.flip(img, axis=(0, 1))
# imgXY = img[::-1, ::-1, :]

# Outputs
cv2_imshow(img)
print("\n\n")
cv2_imshow(imgX)
print("\n\n")
cv2_imshow(imgY)
print("\n\n")
cv2_imshow(imgXY)


```
