# Rotação em 90 graus (sentido anti-horário) e sem uso de escalas

#Referência: https://opencv24-python-tutorials.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_geometric_transformations/py_geometric_transformations.html

```python
import cv2
from google.colab.patches import cv2_imshow
# import numpy as np

im_path = "./imgcourse/Cin.jpg"
img = cv2.imread(im_path)

assert img is not None, "file could not be read, check with os.path.exists()"
print ("shape img = ",img.shape)
rows,cols,_ = img.shape
# cols-1 and rows-1 are the coordinate limits.
M = cv2.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))

# Outputs
cv2_imshow(img)
print("\n\n")
cv2_imshow(dst)



```
